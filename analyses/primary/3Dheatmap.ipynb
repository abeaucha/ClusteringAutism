```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Environment variables
PROJECTPATH = os.getenv('PROJECTPATH')

figdir = os.path.join(PROJECTPATH, "figures", "v3", "figure4")

inputs = [file for file in os.listdir(figdir) if 'opt.csv' in file]
inputs = [os.path.join(figdir, file) for file in inputs]
inputs.sort()

keys = ['HBN_MICe', 'PONDSK_HBN', 'PONDSK_MICe']
sim_opt = {x[0]:pd.read_csv(x[1]) for x in zip(keys, inputs)}
```

```python
from itertools import product
```

```python
matches = [(1,2,1),
         (2,3,4)]

xrange = range(1,3)
yrange = range(1,4)
zrange = range(1,5)
points = list(product(xrange, yrange, zrange))

colors = []; alphas = []
for p in points:
    if p in matches:
        colors.append('red')
        alphas.append(0.8)
    else:
        colors.append('lightgray')
        alphas.append(0.1)

xs = [p[0]-0.5 for p in points]
ys = [p[1]-0.5 for p in points]
zs = [p[2]-0.5 for p in points]

xs = np.array(xs)
ys = np.array(ys)
zs = np.array(zs)

dx = dy = dz = 1

# creating figure 
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
  
# creating the plot 
# plot_geeks = ax.bar3d(xs, ys, zs, 
#                       dx, dy, dz, 
#                       color = colors, 
#                      alpha = 0.2)

for i in range(len(points)): 
    ax.bar3d(xs[i], ys[i], zs[i], dx, dy, dz, color = colors[i], alpha = alphas[i])
  
# setting title and labels 
ax.set_xticks(np.arange(1,3))
ax.set_yticks(np.arange(1,4))
ax.set_zticks(np.arange(1,5))
ax.set_xlabel('POND-SickKids') 
ax.set_ylabel('HBN') 
ax.set_zlabel('MICe') 
ax.margins(0)

outfile = '3Dheatmap_opt.png'
outfile = os.path.join(figdir, outfile)
fig.savefig(outfile, dpi = 300)
plt.close()
```
